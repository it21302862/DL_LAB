```python
import numpy as np
import matplotlib.pyplot as plt

# Create random 100000x1 arrays from exponential, uniform, and normal distributions
exp_data = np.random.exponential(scale=1.0, size=(100000, 1))
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)

# Plot histograms
plt.figure(figsize=(10, 6))
plt.hist(exp_data, density=True, bins=100, histtype="step", color="green", label="exponential")
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="blue", label="uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="red", label="normal")

# Set plot limits and labels
plt.axis([-1, 6, 0, 1.5])
plt.legend(loc="upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5].

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range for x and y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a meshgrid
X, Y = np.meshgrid(x, y)

# Compute Z = X**2 + Y**2
Z = X**2 + Y**2

# Create a 3D plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Plot of Z = X^2 + Y^2')

plt.show()
```

Calculate the pearson standard coefficient and spearman rank correlation for the [HP,
Attack, Defense, Sp. Ark, Sp. Def, Speed] features in the seaborn tutorial.ipynb. Visualize the
results using heatmaps. Make sure the values are displaced in the heatmap as well.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'tips' dataset
df = sns.load_dataset('tips')

# Manually create columns to simulate HP, Attack, Defense, Sp. Atk, Sp. Def, Speed
df['HP'] = df['total_bill']
df['Attack'] = df['tip']
df['Defense'] = df['size']
df['Sp. Atk'] = df['total_bill'] * 1.5
df['Sp. Def'] = df['tip'] * 2
df['Speed'] = df['size'] * 1.2

# Extract relevant features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = df[features]

# Calculate Pearson and Spearman correlation matrices
pearson_corr = data.corr(method='pearson')
spearman_corr = data.corr(method='spearman')

# Plot heatmaps
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', ax=ax[0])
ax[0].set_title('Pearson Correlation')

sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', ax=ax[1])
ax[1].set_title('Spearman Correlation')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/Au_nanoparticle_dataset.csv')

# Create a new dataframe by filtering the specified columns
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]

# Display the first 20 samples of the new dataframe
print(new_df.head(20))

# Calculate the mean, standard deviation, and quartile values for each of the 4 features
stats = new_df.describe()
print(stats)

# Plot the histogram of each of these features in a 1x4 layout
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
features = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
for i, feature in enumerate(features):
    sns.histplot(new_df[feature], ax=axes[i], kde=True)
    axes[i].set_title(f'Histogram of {feature}')
plt.tight_layout()
plt.show()

# Visualize the scatter plots and histograms using pairplot
sns.pairplot(new_df)
plt.show()

# PairGrid with customized plots
g = sns.PairGrid(new_df)
g.map_upper(sns.histplot) # bivariate histogram
g.map_diag(sns.histplot, kde=True) # univariate histogram with KDE
g.map_lower(sns.kdeplot) # bivariate KDE
plt.show()
```
